# Nepal House Price Prediction Using Machine Learning

### Abstract

    Real estate pricing in Nepal remains largely informal and inconsistent,
    making it difficult for buyers, sellers, and investors to make informed
    decisions. This project presents an end-to-end machine learning solution
    for predicting house prices in Nepal using property listing data sourced
    from Hamrobazaar, a leading Nepali real estate platform.
    
    The dataset consists of over 2,200 property listings containing features
    such as land area (in Aana/Ropani), number of bedrooms and bathrooms,
    number of floors, road width and type, house facing direction, parking
    availability, city location, and available amenities. Extensive data
    preprocessing was performed to handle Nepal-specific land measurement
    units — converting Aana, Paisa, and Ropani into square feet — along with
    parsing of unstructured text fields, outlier removal, and log
    transformation of the skewed target variable (SalePrice in NPR).
    
    Exploratory data analysis revealed that house prices are heavily
    right-skewed with a skewness of 32.11, and that land area, number of
    floors, parking availability, and city location (Kathmandu commanding
    the highest premium) are the most influential predictors of price.
    
    Multiple regression models were evaluated including Ridge Regression as
    a baseline and XGBoost as the primary model. XGBoost achieved an R²
    score of 0.63 with an RMSE of approximately NPR 37.6 million on the
    test set, outperforming the Ridge baseline (R² = 0.46). A Scikit-learn
    Pipeline combining preprocessing and model inference was constructed for
    reproducibility and deployment readiness.
    
    The final model was integrated into a Django web application and deployed
    on Himalayan Hosting under the subdomain houseprice.aiwithprakas.com.np,
    enabling users to input property details and receive an instant price
    estimate. This project demonstrates the practical application of machine
    learning to address a real-world problem in the Nepali property market.
    
    Keywords: House Price Prediction, Nepal Real Estate, XGBoost, Regression,
              Machine Learning, Django, Hamrobazaar, Aana to Sqft Conversion

## step 4 Convert Ann to Sqft 
     parse_area - "1-0-0-0 Anna" -> sqft
     Nepal ma land Aana/Ropani ma mapinchha — model lai sqft chahinchha. Conversion rules:

     1 Ropani = 16 Aana  = 5476 sqft
    1 Aana   = 4 Paisa  = 342.25 sqft
    1 Paisa  = 4 Dam    = 85.56 sqft
    1 Dam    =            21.39 sqft

    Format "1-0-0-0 Aana" ko matlab:
        1 Ropani, 0 Aana, 0 Paisa, 0 Dam
            = 1×16×342.25 = 5476 sqft

        re.sub(r'(\d)\.(\d)', r'\1-\2', s) le "3.1.0 Aana" jasto dot-separated format lai "3-1-0 Aana" ma convert garcha — kinabhane dataset ma inconsistent format thiyo.

```python
df.Area
```

```python
import re
def parse_area(s):
    #-----------------------------------------
    # parse_area - "1-0-0-0 Anna" -> sqft
    # Nepal ma land anna/Ropani ma napinxa  
    # hamile teslai sqrt ma change garnu parxa 
    # -----------------------------------------------
    s = str(s).strip()
    # Normalize dot-separated numbers (e.g. "3.1.0 Aana" → "3-1-0 Aana")
    s2 = re.sub(r'(\d)\.(\d)', r'\1-\2', s)
 
    # R-A-P-D Aana (4 parts)
    m = re.match(r'^(\d+)-(\d+)-(\d+)-(\d+)\s*/?\s*[\d.]*\s*Aana', s2, re.I)
    if m:
        r, a, p, d = int(m.group(1)), int(m.group(2)), int(m.group(3)), int(m.group(4))
        return r * 16 * 342.25 + a * 342.25 + p * 85.56 + d * 21.39
    # R-A-P Aana (3 parts)
    m = re.match(r'^(\d+)-(\d+)-(\d+)\s*/?\s*[\d.]*\s*Aana', s2, re.I)
    if m:
        r, a, p = int(m.group(1)), int(m.group(2)), int(m.group(3))
        return r * 16 * 342.25 + a * 342.25 + p * 85.56
    # X Aana / X ana Aana
    m = re.match(r'^([\d.]+)\s*(?:ana\s*|aana\s*)?Aana', s, re.I)
    if m:
        try: return float(m.group(1)) * 342.25
        except: return np.nan
    # R-A-P-D Ropani
    m = re.match(r'^(\d+)-(\d+)-(\d+)-(\d+)\s*Ropani', s2, re.I)
    if m:
        r, a, p, d = int(m.group(1)), int(m.group(2)), int(m.group(3)), int(m.group(4))
        return r * 5476 + a * 342.25 + p * 85.56 + d * 21.39
    # X Ropani
    m = re.match(r'^([\d.]+)\s*Ropani', s, re.I)
    if m: return float(m.group(1)) * 5476
    # X Sq. Feet
    m = re.match(r'^([\d.]+)\s*Sq', s, re.I)
    if m: return float(m.group(1))
    return np.nan
 
df['Area_sqft'] = df['Area'].apply(parse_area)
```

## Step 5 Road Road Width
    20 feet / backtopped bata 20 matra nikalera 
    Meter aayo bhane x 3.281 gari feet ma convert garne 



```python
df['Road Width']
```

```python
# --- Road Width: "20 Feet / Blacktopped" → 20 ---
def parse_road_width(rw):
    m = re.search(r'([\d.]+)\s*(Feet|Meter|feet|meter)', str(rw))
    if m:
        val = float(m.group(1))
        if 'meter' in m.group(2).lower():
            val *= 3.281  # meter to feet
        return val
    return np.nan
```

```python
df['RoadWidth_ft'] = df['Road Width'].apply(parse_road_width)
```

    road_rank = {'Blacktopped':4, 'Concrete':3, 'Paved':2, 'Gravelled':1}

```python
# --- Road Type: string → rank ---
road_rank = {'Blacktopped': 4, 'Concrete': 3, 'Paved': 2, 'Gravelled': 1}
def parse_road_type(rt):
    for k, v in road_rank.items():
        if k.lower() in str(rt).lower():
            return v
    return 0
```

```python
df['RoadType_num'] = df['Road Type'].apply(parse_road_type)
```

```python
# step 1
# Missing ra Non-missing grop bich ma Price ma difference xa ki xaina ?
missing_floors = dataset[dataset['Floors'].isnull()]['Price']
non_missing_floors = dataset[dataset['Floors'].notnull()]['Price']
print("missing_floors_mean:",missing_floors.mean())
print("non_missing_floors_mean: ", non_missing_floors.mean())
```

```python
# step 2 
from scipy.stats import mannwhitneyu
stas,p = mannwhitneyu(missing_floors, non_missing_floors)
if p < 0.05:
    print( "Significant Difference -> missing important - add indicator, missing हुनु र Price बीच सम्बन्ध छ (missing important छ) ")
else :
    print("Not significant - missing random-Drop/fill")
    
```

##### floor Missing huda price High xa 
・ yo difference real ho ki chance le matra ayeko ho test garam

```python
# Example DataFrame
from scipy.stats import ttest_ind
df = dataset.copy()
columns_to_check = ['Floors', 'Year', 'Road Type']  # Missing check गर्ने columns
target_col = 'Price'                        # Target column

# Result store गर्ने list
results = []

for col in columns_to_check:
    missing_group = df[df[col].isnull()][target_col]
    non_missing_group = df[df[col].notnull()][target_col]
    
    # Check if both groups have enough data
    if len(missing_group) > 1 and len(non_missing_group) > 1:
        t_stat, p_value = mannwhitneyu(missing_group, non_missing_group)
        missing_avg = missing_group.mean()
        non_missing_avg = non_missing_group.mean()
        diff = abs(missing_avg - non_missing_avg)
        
        results.append({
            'Column': col,
            'Missing Avg': missing_avg,
            'Non-Missing Avg': non_missing_avg,
            'Diff': diff,
            'T-stat': t_stat,
            'P-value': p_value
        })
    else:
        results.append({
            'Column': col,
            'Missing Avg': np.nan,
            'Non-Missing Avg': np.nan,
            'Diff': np.nan,
            'T-stat': np.nan,
            'P-value': np.nan
        })

# Results देखाउने
results_df = pd.DataFrame(results)
print(results_df)
    
```

```python
for col, p_value in zip(results_df['Column'], results_df['P-value']):
    if p_value < 0.05:
        print(f"{col} -> Significant Difference: missing important, add indicator")
    else:
        print(f"{col} -> Not Significant: drop or impute")
```
